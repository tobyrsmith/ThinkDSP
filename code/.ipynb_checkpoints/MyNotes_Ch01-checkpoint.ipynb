```python
from __future__ import print_function, division

from IPython.display import Audio

from IPython.html.widgets import interact, fixed
from IPython.html import widgets

import thinkdsp
import thinkplot

import numpy as np

%matplotlib inline
```

```python
I_sig = thinkdsp.SinSignal(freq=261.63, amp=1.0, offset=0)    #c4
III_sig = thinkdsp.SinSignal(freq=329.63, amp=0.5, offset=0)   #e4
VII_sig = thinkdsp.SinSignal(freq=493.88, amp=0.25, offset=0)   #b4
```

```python
#I_sig.plot()
```

```python
mix0 = I_sig
mix1 = I_sig + III_sig
mix2 = I_sig + III_sig + VII_sig
```

```python
wave0 = mix0.make_wave(duration=1.0, start=0, framerate=11025)
wave0.make_audio()
```

```python
wave1 = mix1.make_wave(duration=1.0, start=0, framerate=11025)
wave1.make_audio()
```

```python
wave2 = mix2.make_wave(duration=1.0, start=0, framerate=11025)
wave2.make_audio()
```

```python
period = mix2.period
segment = wave2.segment(start=0, duration=period*10)
segment.plot()
```
